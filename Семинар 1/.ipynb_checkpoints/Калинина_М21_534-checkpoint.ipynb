```python
negative = ['проблем', 'ненавижу', 'сомнительный', 'минус']
```

```python
for i in range(len(negative)):
  print([x.find(negative[i]) for x in reviews if x.find(negative[i])!= -1])
```

```python
def checkWords(review):
  l = len([review.find(word) for word in negative if review.find(word)!= - 1])
  return l if l > 0 else 0
```

```python
sum([checkWords(review) for review in reviews])/ len(reviews)
```

```python
reviews = []
for i in range(0, 3, 1): #i равняется 1, 2, 3, 4
  page = requests.get('https://irecommend.ru/content/gosuslugi?page={}'.format(str(i))).text
  soup = BeautifulSoup(page, 'html.parser')
  reviews.extend([x.text for x in soup.find('ul', class_ = 'list-comments').find_all('li')])
```

```python
reviews = []
for i in range(0, 4, 1): #i равняется 1, 2, 3, 4
  page = requests.get('https://irecommend.ru/content/gosuslugi?page={}'.format(str(i))).text
  soup = BeautifulSoup(page, 'html.parser')
  reviews.extend([[len(x.find('div', class_ = "starsRating").find_all('div', class_ = "on")), x.find('div', class_ = "created").text, x.text] for x in soup.find('ul', class_ = 'list-comments').find_all('li')])
  print(len(reviews))
```

```python
l = [[x[1][-4:], x[1][-7:-5], x[0]] for x in reviews]
l.sort()
y_pred, m_pred = l[0][0], l[0][1]
sum_, count_ = 0,0
stars_list = []
for y, m, s in l:
  if y == y_pred and m == m_pred:
    sum_+=s
    count_+=1
    #print(sum_, count_)
  else:
    stars_list+=[[y_pred, m_pred, sum_/ count_]]
    sum_, count_=s,1
  y_pred, m_pred = y, m
stars_list
```
