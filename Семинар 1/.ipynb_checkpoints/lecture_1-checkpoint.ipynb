```python
reviews=[x.text.replace('\n', '').lower() for x in soup.find('ul', class_ = 'list-comments').find_all('li')]
reviews
```

```python
len(reviews)
```

```python
bad_words = ['проблем','минус','ненависть','невозможно', 'ужас']
```

```python
word = bad_words[4]
len([1 for x in reviews if x.replace('\n','').lower().find(word)!=-1])/len(reviews)*100
```

```python
for i in range(len(bad_words)):
  print([x.find(bad_words[i]) for x in reviews if x.find(bad_words[i])!= -1])
```

```python
def check_words(review):
  l=len([review.lower().find(word) for word in bad_words if review.lower().find(word)!=-1])
  return 1 if l > 0 else 0
```

```python
for i in range(len(reviews)):
    res = check_words(reviews[i])
    if res != 0:
        print(i, res)
```
