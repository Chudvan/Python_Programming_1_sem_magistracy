```python
from flask import Flask, render_template_string, request
import flask
from flask_ngrok import run_with_ngrok
```

```python
app = Flask(__name__)
run_with_ngrok(app)

@app.route("/", methods=['GET', 'POST'])
def hello():
    if request.method == 'GET':
        con = sqlite3.Connection('vacancies.db')
        avrg = pd.read_sql('select * from "employers";', con=con).to_json()
        return render_template_string(avrg)
    else:
        con = sqlite3.Connection('vacancies.db')
        city = request.get_json()['city']
        avrg_by_city = pd.read_sql('''select avg(salary) from "vacancies" where area=='{}';'''.format(city), con=con).to_json()
        return avrg_by_city

if __name__=='__main__':
    app.run()
```
