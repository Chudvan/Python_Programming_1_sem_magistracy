# Создание и заполнение таблиц

## Создание таблиц

```python
import sqlite3
```

```python
connection = sqlite3.connect('./vacancies.db')
```

```python
create_employer_table_query = """
create table employer (
  id integer PRIMARY KEY,
  name text
)
"""
connection.execute(create_employer_table_query)
connection.commit()
```

```python
create_area_table_query = """
create table area (
  id integer PRIMARY KEY,
  name text,
  parent_id integer,
  FOREIGN KEY (parent_id) REFERENCES area(id)
)
"""
connection.execute(create_area_table_query)
connection.commit()
```

```python
create_vacancy_table_query = """
create table vacancy (
  id integer PRIMARY KEY,
  name text,
  salary real,
  area_id,
  employer_id,
  FOREIGN KEY (area_id) REFERENCES area(id),
  FOREIGN KEY (employer_id) REFERENCES employer(id)
)
"""
connection.execute(create_vacancy_table_query)
connection.commit()
```

```python
currency_rate = {
    "RUR": 1,
    "USD": 73,
    "EUR": 85,
    "KZT": 0.17,
    "UZS": 0.0068,
    "UAH": 3,
    "BYR": 30,
    "AZN": 43,
    "KGS": 1.17
}

def calc_salary(salary):
  if salary is None:
    return None

  if salary['from'] is None:
    avg = salary['to']
  elif salary['to'] is None:
    avg = salary['from']
  else:
    avg = (salary['from'] + salary['to']) / 2
  
  avg *= currency_rate[salary['currency']]
  if salary['gross']:
    avg *= 0.87
  return avg
```

```python
#Записываем все возможные area в таблицу area
def process_area(areas, connection):
  for area in areas:
    connection.execute("INSERT OR IGNORE INTO area VALUES(?, ?, ?)", (area['id'], area['name'], area['parent_id']))
    if len(area['areas']) > 0:
      process_area(area['areas'], connection)

process_area(areas_res, connection)
connection.commit()
```

```python
def insert_employers(df,connection):
  for i, x in df.iterrows():
    employer = x['employer']
    connection.execute("insert or ignore into employer values(?, ?)", (employer['id'], employer['name']))
  connection.commit()

insert_employers(data, connection)
connection.commit()
```

```python
def get_vaсancies_by_name(vacancy_name):
  response = requests.get("https://api.hh.ru/vacancies", 
              params={
                  "text": vacancy_name,
                  "per_page": "100",
              }).json()
  df = pd.DataFrame(response["items"]);
  df = df.loc[:,['id', 'name', 'area','salary','employer']].\
  assign(area=df.area.apply(lambda x: x['id'])).\
  assign(salary=df.salary.apply(calc_salary))
  return df
```

```python
app = Flask(__name__)
run_with_ngrok(app)
# print(app.config)
app.config['JSON_AS_ASCII'] = False

@app.route("/groups",  methods=['GET'])
def get_grouped_vacancies():
  connection = sqlite3.connect('./vacancies.db')
  response = '';
  vacancies = pd.read_sql('select vacancy.name, vacancy.salary, area.name as area from vacancy join area on vacancy.area_id = area.id', con=connection);
  groups = ["Продавец", "Менеджер", "Инженер", "Курьер"]
  for group in groups:
    response += f'<h1>{group}</h1>'
    df = vacancies[vacancies.apply(lambda x: group.lower() in x['name'].lower(), axis=1)] #Фильтрация df по group
    response += df.groupby(['area']).agg({'salary': ['min', 'max', 'mean', 'median']}).to_html()
  connection.close()
  return response;

@app.route("/vacancies",  methods=['POST'])
def add_vacancies():
  connection = sqlite3.connect('./vacancies.db')
  vacancies = get_vaсancies_by_name(request.json['vacancy_name'])
  insert_employers(vacancies, connection)
  insert_vacancies(vacancies, connection)
  connection.close()
  return '', 204

@app.route("/vacancies",  methods=['GET'])
def get_all_vacancies():
  connection = sqlite3.connect('./vacancies.db')
  vacancies = pd.read_sql('select * from vacancy', con=connection)
  vacancies = vacancies.where(pd.notnull(vacancies), None).to_dict(orient="records"); #Список записей-словарей вакансий
  employers = pd.read_sql('select * from employer', con=connection).to_dict(orient="records"); #Список записей-словарей работодателей
  connection.close()
  return {"employers": employers, "vacancies": vacancies};


if __name__ == "__main__":
    app.run()
```

```python
def get_vanacies_by_name_and_region(vacancy_name, region_id):
  response = requests.get("https://api.hh.ru/vacancies", 
              params={
                  "text": vacancy_name,
                  "per_page": "100",
              }).json()
  df = pd.DataFrame(response["items"]);
  df = df.loc[:,['id', 'name', 'area','salary','employer']].\
  assign(area=df.area.apply(lambda x: x['id'])).\
  assign(salary=df.salary.apply(calc_salary))
  return df

class VacanciesStorage:  

    def __init__(self, dbfile):  
      self.connection = sqlite3.connect(dbfile);
      create_vacancy_table_query = """
      create table if not exists vacancy (
        id integer PRIMARY KEY,
        name text,
        salary real,
        area_id,
        employer_id,
        FOREIGN KEY (area_id) REFERENCES area(id),
        FOREIGN KEY (employer_id) REFERENCES employer(id)
      )
      """
      create_employer_table_query = """
      create table if not exists employer (
        id integer PRIMARY KEY,
        name text
      )
      """
      create_area_table_query = """
      create table if not exists area (
        id integer PRIMARY KEY,
        name text,
        parent_id integer,
        FOREIGN KEY (parent_id) REFERENCES area(id)
      )
      """
      self.connection.execute(create_area_table_query)
      self.connection.execute(create_employer_table_query)
      self.connection.execute(create_vacancy_table_query)
      self.connection.commit()

      areas_res = requests.get("https://api.hh.ru/areas").json();
      connection = sqlite3.connect('/content/vacancies.db')
      process_area(areas_res, self.connection)
      self.connection.commit()

    def load_vacancies(self, vacancy_name, area_name):
      res = pd.read_sql(f"select id from area where name = '{area_name}'", con=self.connection)
      area_id = res['id'].iloc[0]
      df = get_vanacies_by_name_and_region(vacancy_name, area_id)
      insert_employers(df, self.connection)
      insert_vacancies(df, self.connection)

    def get_all_vacancies(self):  
      print("hah")
```
