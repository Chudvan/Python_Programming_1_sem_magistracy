```python
app = Flask(__name__)
run_with_ngrok(app)  # Start ngrok when app is run

@app.route("/")
def index():
  return render_template_string('''<h3>ДЗ№2</h3><br><br><br><a href="/show1"; style="color: red; font-size:16pt; font-family:Calibri";>Вывод вакансий по маркетингу</a><br><br><br><a href="/show2" style="color: red; font-size:16pt; font-family:Calibri";>Таблица всех вакансий</a>''')

@app.route("/show1", methods=['GET','POST'])
def show1():
  conn=sqlite3.Connection('./new.db')
  if request.method=='GET':
      text=pd.read_sql('''select * from "vacancies" where name LIKE '%по маркетингу';''',conn).to_html()
      return render_template_string(text)
  else:
      sql='''select 
          avg(sal) from "vacancies" 
          where 
          name LIKE '%{}';'''.format(request.get_json()['vacancy'])
      text=pd.read_sql(sql,conn).values[0][0]
      context={'vacancy':str(text)}
      return render_template_string(json.dumps(context))

@app.route("/show2", methods=['GET','POST'])
def show2():
  conn=sqlite3.Connection('./new.db')
  if request.method=='GET':
      text = '<p>Таблица вакансий</p>'
      text += '<h3 style="color: red; font-size:20pt; font-family:Calibri;">Введите выше какой ЗП выводить вакансии.</h3><br><form action="/show2" class="test__form" method="POST"><input type="text" name="word_" id="word" placeholder="Введите ЗП"><br><input type="submit" name="but" value="Отправить"><br><br><a href="/">Назад</a>'
      return render_template_string(text)
  else:
      word = request.form['word_']
      print(word)
      sql=f'''select 
          * from "vacancies" 
          where 
          sal > {word};'''
      text=pd.read_sql(sql,conn).to_html()
      conn.close()
      return(render_template_string(text))

if __name__ == '__main__':
    app.run()
```
