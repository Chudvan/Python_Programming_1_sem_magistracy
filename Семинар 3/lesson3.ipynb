```python
app = Flask(__name__)
run_with_ngrok(app)

@app.route("/", methods=['GET', 'POST'])
def hello():
    if request.method == 'GET':
        conn = sqlite3.Connection('./vacancies.db')
        avrg = pd.read_sql('select avg(salary) from "vacancies";', con=conn).to_json()
        return render_template_string(avrg)
    else:
#         print(request.get_json())
        conn = sqlite3.Connection('./vacancies.db')
        city = request.get_json()['city']
        avrg_by_city = pd.read_sql('''select avg(salary) from "vacancies" where area=='{}';'''.format(city), con=conn).to_json()
        return avrg_by_city

if __name__=='__main__':
    app.run()
```
