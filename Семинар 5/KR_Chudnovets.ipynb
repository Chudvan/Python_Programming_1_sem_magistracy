```python
import requests
```

```python
import pandas as pd
```

```python
d = {}
l = []
text_list = ['data+analyst', 'dataANDengineer']
for i, text in enumerate(text_list):
  for j in range(5):
    items = requests.get('https://api.hh.ru/vacancies?text={}&per_page=100&page={}'.format(text, j)).json()['items']
    for item in items:
      d['id'] = item['id']
      d['name'] = item['name']
      d['salary'] = item['salary']
      d['area'] = item['area']
      d['published_at'] = item['published_at']
      l.append(d.copy())
  if i == 0:
    df_1 = pd.DataFrame(l)
    l = []
  elif i == 1:
    df_2 = pd.DataFrame(l)
    l = []
```
