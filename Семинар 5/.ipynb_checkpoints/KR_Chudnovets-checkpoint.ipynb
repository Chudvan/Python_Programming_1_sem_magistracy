```python
import requests
```

```python
import pandas as pd
```

```python
d = {}
l = []
text_list = ['data+analyst', 'dataANDengineer']
for i, text in enumerate(text_list):
  for j in range(5):
    items = requests.get('https://api.hh.ru/vacancies?text={}&per_page=100&page={}'.format(text, j)).json()['items']
    for item in items:
      d['id'] = item['id']
      d['name'] = item['name']
      d['salary'] = item['salary']
      d['area'] = item['area']
      d['published_at'] = item['published_at']
      l.append(d.copy())
  if i == 0:
    df_1 = pd.DataFrame(l)
    l = []
  elif i == 1:
    df_2 = pd.DataFrame(l)
    l = []
```

```python
df_3 = pd.concat([df_1, df_2], axis=0)
df_3
```

```python
df_3['area'] = df_3.area.apply(lambda x: x['name'])
df_3
```

```python
import datetime
format_ = '%Y-%m-%dT%H:%M:%S%z'
date_string = list(df_3.published_at)[0]
cur_date = datetime.datetime.strptime(date_string, format_)

cur_date
```
